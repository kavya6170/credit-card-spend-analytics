```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import current_timestamp, input_file_name, lit
from pyspark.sql.types import *
```

```python
print("Starting Bronze Layer ETL...")
print("Serverless compute is provisioning automatically...")
```

```python
bronze_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "false")  # IMPORTANT: raw data only
    .csv("/Volumes/workspace/default/credit_data/credit_card_transactions.csv")
)

print("Bronze data loaded successfully")
```

```python
bronze_df=bronze_df.withColumnRenamed(bronze_df.columns[0], "transaction_id")
```

```python
bronze_df.columns
```

```python
bronze_df = bronze_df.withColumn(
    "ingestion_timestamp",
    current_timestamp()
)
```

```python
bronze_df.show(5, truncate=False)
bronze_df.printSchema()
```

```python
bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_credit_transactions")

print("Bronze table created successfully!")
```
